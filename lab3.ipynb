```python
import matplotlib.pyplot as plt
%matplotlib inline

#from __future__ import unicode_literals
from collections import Counter

import matplotlib
matplotlib.rc('font',family='NanumGothic') 

import csv
f = open('data.csv')

A=[]
B=[]

for row in csv.reader(f):
    A.append(float(row[0]))
    B.append(float(row[1]))

f.close()
```

```python
def make_simple_line_chart():
    """ 1 """
    
    
    # create a line chart, years on x-axis, gdp on y-axis
    plt.plot(A, B, color='green', marker='v', linestyle='solid')
    # makrer, linestyle등 표시 방법은 http://matplotlib.org/api/lines_api.html 참조

    # add a title
    plt.title("")

    # add a label to the y-axis
    plt.ylabel("")
    plt.show()
    
make_simple_line_chart()
```

```python
def make_simple_bar_chart():
    """ 2 """

    

    # bars are by default width 0.8, so we'll add 0.1 to the left coordinates
    # so that each bar is centered
    xs = [i + 0.1 for i, _ in enumerate(A)]
    # xs = [0.1, 1.1, 2.1, 3.1, 4.1], 일반적은로 _ 변수는 쓰지 않는 변수를 할당할 때 사용함

    # plot bars with left x-coordinates [xs], heights [num_oscars]
    plt.bar(xs, B)
    plt.ylabel("")
    plt.title("")

    # label x-axis with movie names at bar centers
    plt.xticks([i + 0.5 for i, _ in enumerate(A)], A)

    plt.show()

make_simple_bar_chart()
```

```python
def make_several_line_charts():
    """ 4 """
    
    total_error  = [x + y for x, y in zip(A,B)]
    # [257, 130, 68, 40, 32, 40, 68, 130, 257]

    xs = range(len(A))

    # we can make multiple calls to plt.plot
    # to show multiple series on the same chart
    plt.plot(xs, A,     'g-',  label='A')    # green solid line
    plt.plot(xs, B, 'r-.', label='B')      # red dot-dashed line
    plt.plot(xs, total_error,  'b:',  label='total error') # blue dotted line

    # because we've assigned labels to each series
    # we can get a legend for free
    # loc=9 means "top center"
    plt.legend(loc=9)
    plt.xlabel("")
    plt.title("")
    plt.show()

make_several_line_charts()
```
