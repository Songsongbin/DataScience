```python
#예제1
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=[2,4,6,8]
y=[81,93,91,97]

mx=np.mean(x)
my=np.mean(y)
print("x의 평균값:",mx)
print("y의 평균값:",my)

divisor=sum([(mx-i)**2 for i in x])

def top(x,mx,y,my):
    d=0
    for i in range(len(x)):
        d+=(x[i]-mx)*(y[i]-my)
        return d
dividend=top(x,mx,y,my)

print("분모:",divisor)
print("분자:",dividend)

a=dividend/divisor
b=my-(mx*a)

print("기울기 a =",a)
print("y절편 b =",b)

x1=np.arange(0,10,1)
y1=[(a*num+b)for num in x1]
plt.scatter(x,y,color="blue")
plt.plot(x1,y1,color="red")
plt.show()
```

```python
#예제2
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ab=[3,76]
data=[[2,81],[4,93],[6,91],[8,97]]
x=[i[0] for i in data]
y=[i[1] for i in data]

def predict(x):
    return ab[0]*x+ab[1]

def rmse(p,a):
    return np.sqrt(((p-a)**2).mean())

def rmse_val(predict_result,y):
    return rmse(np.array(predict_result),np.array(y))

predict_result=[]

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간=%.f,실제점수=%.f,예측점수=%.f"%(x[i],y[i],predict(x[i])))

print("rmse 최종값:"+str(rmse_val(predict_result,y)))
print("기울기 a=",ab[0])
print("y절편 b=",ab[1])

plt.plot(x,predict_result,':',color="green")
plt.show()
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ab=[3,76]
data=[[2,81],[4,93],[6,91],[8,97]]
npdata=np.array(data)
x=[i[0] for i in data]
y=[i[1] for i in data]

def predict(x):
    return ab[0]*x+ab[1]

def rmse(p,a):
    return np.sqrt(((p-a)**2).mean())

def rmse_val(predict_result,y):
    return rmse(np.array(predict_result),np.array(y))

def step_gradient(w_current,b_current,points,learning_rate):
    w_gradient=0
    b_gradient=0
    N=float(len(points))
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        w_gradient+=-(2/N)*x*(y-(w_current *x+b_current))
        b_gradient+=-(2/N)*(y-(w_current *x+b_current))
    new_w = w_current - (learning_rate* w_gradient)
    new_b = b_current - (learning_rate* b_gradient)
    return [new_w,new_b]

predict_result=[]

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간=%.f,실제점수=%.f,예측점수=%.f"%(x[i],y[i],predict(x[i])))

print("rmse 최종값:"+str(rmse_val(predict_result,y)))
step_gradient(ab[0],ab[1],npdata,0.03)
print("기울기 a=",ab[0])
print("y절편 b=",ab[1])

plt.plot(x,predict_result,':',color="green")
plt.show()
```
